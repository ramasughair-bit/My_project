```python
import hashlib
import string


def check_password(password):
    reasons = []

    if len(password) < 8:
        reasons.append("Less than 8 characters")

    if not any(c.isupper() for c in password):
        reasons.append("No uppercase letter")

    if not any(c.islower() for c in password):
       reasons.append("No lowercase letter")

    if not any(c.isdigit() for c in password):
       reasons.append("No number")

    if not any(c in string.punctuation for c in password):
       reasons.append("No special character")

    if reasons:
        print("Password is weak")
        print("Reasons:")
        for r in reasons:
            print("-", r)
    else:
        print("Password is strong")

        hash_value = hashlib.sha256(password.encode()).hexdigest()
        print("Hash:", hash_value)



def check_jwt(secret):
    reasons = []
    weak_values = ["admin", "123456", "password", "secret", "jwtsecret"]

    if len(secret) < 16:
        reasons.append("Less than 16 characters")

    if not any(c.isupper() for c in secret):
        reasons.append("No uppercase letter")

    if not any(c.islower() for c in secret):
        reasons.append("No lowercase letter")

    if not any(c.isdigit() for c in secret):
        reasons.append("No number")

    if not any(c in string.punctuation for c in secret):
        reasons.append("No special character")

    if secret.lower() in weak_values:
        reasons.append("Common weak value")

    if reasons:
        print("JWT Secret is weak")
        print("Reasons:")
        for r in reasons:
            print("-", r)
    else:
        print("JWT Secret is strong")


password = input("Enter Password: ")
check_password(password)

print()

jwt_secret = input("Enter JWT Secret: ")
check_jwt(jwt_secret)
```
